```python
import time
import tensorflow as tf
import gym 
import random 
```

```python
env= gym.make("CartPole-v1")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
from rl.agents import DQNAgent
from rl.memory import SequentialMemory
from rl.policy import BoltzmannQPolicy
```

```python
model = Sequential()
model.add(Flatten(input_shape=(1,4)))
model.add(Dense(48, activation="relu"))
model.add(Dense(24, activation="relu"))
model.add(Dense(2,activation="linear"))

model.summary()
```

```python
policy = BoltzmannQPolicy()
memory = SequentialMemory(limit=50000 , window_length=1)
robot = DQNAgent(model=model , policy=policy , memory=memory , nb_actions=2,
                     nb_steps_warmup=10, target_model_update=1e-2 )

robot.compile(Adam(lr=1e-3) , metrics=['mae'])
```

```python
robot.fit(env=env, nb_steps=5000, visualize=False, verbose=True)
```

```python
out = robot.test(env=env, nb_episodes=10, visualize=True, verbose=True)
print(np.mean(out.history['episode_reward']))
```
