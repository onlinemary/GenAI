```python
import time
import tensorflow as tf
import gym 
import random 
```

```python
env= gym.make("CartPole-v1")

env.reset()

run = True
st =0
while run:
    time.sleep(0.01)
    env.render()
    act= random.randint(0,1)
    CartP ,CartV ,PoleA, PoleV = env.step(act)
    if CartP[0] < -1:
        act = 1
        env.step(act)
    elif CartP[0] > 1:
        act = 0
        env.step(act)
    print(CartP[0])
    st+=1
    if st > 3000 :
        run = False
        st = 0
    
env.close()
```
